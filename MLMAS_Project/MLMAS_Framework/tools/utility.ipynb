## Returns the id of the hero vehicle

```python
def get_hero_vehicle(client, recorder_file):
    try:
        act_list = client.show_recorder_file_info(recorder_file, False).splitlines()
        for i in range(len(act_list)):
                if "role_name = hero" in act_list[i]:
                    s = act_list[i-1]
                    s = s[7:s.find(':')]
                    # print("Hero found at: ", int(s))
                    return int(s)

    finally:
        pass
```

## Returns the time from infraction coordinates

```python
json_file = prj_dir + "/results/MLMAS_LAV_results.json"

list_of_infractions = {}
meta_data = {}
result = []
```

```python
# read the json file
with open(json_file) as f:
    data = json.load(f)
    main_list = data['_checkpoint']['records']

    for i in main_list:
        list_of_infractions[i['index']] = i['infractions']
        meta_data[i['index']] = i['meta']

len(list_of_infractions)
```

```python
# avoided infractions
avoid_infractions = ['outside_route_lanes', 'route_timeout']

def extract_coordinates_from_infraction(infraction):
    loc = []
    for inf in infraction:
        x_ind = inf.find("x=")
        y_ind = inf.find("y=")
        z_ind = inf.find("z=")

        x = inf[x_ind+2:y_ind-2]
        y = inf[y_ind+2:z_ind-2]
        z = inf[z_ind+2:inf.find(")")]

        loc.append(
                {'x':x,
                'y':y,
                'z':z,
                'inf':inf})

    return loc

def get_infraction_coordinates(infraction_param):
    map_waypoints = []
    for key in infraction_param.keys():
        if len(infraction_param[key]) > 0 and key not in avoid_infractions:
            # print(infraction_param[key])
            loc = extract_coordinates_from_infraction(infraction_param[key])
            map_waypoints.append(loc)
        # elif key in avoid_infractions:
        #     loc = [{'x':0,
        #         'y':0,
        #         'z':0,
        #         'inf':key}]
        #     map_waypoints.append(loc)

    flattened_coordinates =  [coordinate for sublist in map_waypoints for coordinate in sublist]

    return [{'loc': carla.Location(float(coordinate['x']), float(coordinate['y']), float(coordinate['z'])),
                    'inf': coordinate['inf']} for coordinate in flattened_coordinates]

def calculate_time_for_infraction(client, recorder_file, infraction_param, meta_param, distance_threshold=5.0, debug=False):
    time = []
    # Get the infraction coordinates
    infraction_coordinates = get_infraction_coordinates(infraction_param)
    world = client.get_world()
    # Get the actor ID of the hero vehicle from the recorded file
    actor_id = get_hero_vehicle(client, recorder_file)
    actor = world.get_actor(actor_id)
    # loop to continuously check the actor's location during replay   
    while len(infraction_coordinates) > 0:
        # Get the actor's live location
        actor_location = actor.get_location()
        if world.get_snapshot().timestamp.elapsed_seconds > (meta_param['duration_game'] + 10.0):
            if len(infraction_coordinates) > 0:
                for infraction in infraction_coordinates:
                    time.append({'time': 'unavailable', 'inf': infraction['inf']})
            break
        # Check if the actor is near any of the infraction coordinates
        for index_c, coordinate in enumerate(infraction_coordinates):
            #if coordinate['inf'] not in avoid_infractions:
                distance = actor_location.distance(coordinate['loc'])
                if distance <= distance_threshold:
                    current_time = world.get_snapshot().timestamp.elapsed_seconds
                    #append the time of the infraction and the infraction message
                    time.append({'time': current_time, 'inf': coordinate['inf']})
                    if debug:
                        print("Inference message:", coordinate['inf'], "at time:", current_time)

                    # remove the infraction from the list
                    infraction_coordinates.pop(index_c)    
            #else:
           #     time.append({'time': 'unavailable', 'inf': coordinate['inf']}) 

    return time

def start_simulation_infraction(recorder_file, infraction_param, meta_param, time_factor=1.0, distance_threshold=5.0, debug=False):
    client = carla.Client(HOST, PORT)
    client.set_timeout(60.0)
    client.reload_world()

    # Set the time factor for the replay
    client.set_replayer_time_factor(time_factor)

    actor_id = get_hero_vehicle(client, recorder_file)

    replay = client.replay_file(recorder_file,
                       time_start=0.0,
                       duration=0.0,
                       follow_id = actor_id)
    
    print(replay)

    return calculate_time_for_infraction(client, recorder_file, infraction_param, meta_param, distance_threshold=distance_threshold, debug=debug)
```

```python
# create the driver log files
# create a function that accepts a list of numbers and removes the numbers that are having difference 1, keep the smaller number
def remove_consecutive_numbers(list_of_numbers):
    # sort the list of numbers
    list_of_numbers.sort()

    # loop through the list of numbers
    for i in range(len(list_of_numbers)-1, 0, -1):
        # check if the difference between the numbers is 1
        if list_of_numbers[i] - list_of_numbers[i-1] == 1:
            # remove the number at index i
            list_of_numbers.pop(i)


    return list_of_numbers


# a function that returns sub dataframes of the main dataframe with indexes of the start and end from the abocve list
def get_sub_dataframes(df, list_of_indexes):
    # create an empty list to store the sub dataframes
    list_of_sub_dataframes = []

    # loop through the list of indexes
    for i in range(len(list_of_indexes)-1):
        # get the start and end index
        start_index = list_of_indexes[i]
        end_index = list_of_indexes[i+1]

        # print(start_index, end_index)

        # get the sub dataframe
        sub_df = df.iloc[start_index:end_index]


        # append the sub dataframe to the list
        list_of_sub_dataframes.append(sub_df)

        # if the end index is the last index
        if end_index == list_of_indexes[-1]:
            # get the sub dataframe from end index to the end of the dataframe
            sub_df = df.iloc[end_index:]
            # print(end_index)
            # append the sub dataframe to the list
            list_of_sub_dataframes.append(sub_df)

    return list_of_sub_dataframes


# create a function that saves each sub dataframe to a csv file
def save_sub_dataframes_to_csv(list_of_sub_dataframes):
    # loop through the list of sub dataframes
    for i in range(len(list_of_sub_dataframes)):
        # get the sub dataframe
        sub_df = list_of_sub_dataframes[i]
        

        
        # save the sub dataframe to a csv file
        sub_df.to_csv(f"prj_dir + '/results/records/MLMAS_LAV/driver_{i}.csv", index=False)
        print 
```

```python
def mergeTwoLeaderbordMetricsResults(title1, title2, df1, df2):
    score_columns = df1.columns.values[0:3]
    df_s1 = df1[score_columns]
    df_s1.loc[:,2:3] = df_s1[score_columns[2]].values*100
    df_s2 = df2[score_columns]
    df_s2.loc[:,2:3] = df_s2[score_columns[2]].values*100
    
    collisions_columns = df1.columns.values[3:6]
    
    road_columns = df1.columns.values[6:9]
    
    other_columns = df1.columns.values[9:]
    
    final_df_s = pd.DataFrame({title1: df_s1.values[0]
                  , title2: df_s2.values[0]
                  , 'Results_Higher_is_Better': score_columns}) 
    
    final_df_c = pd.DataFrame({title1: df1[collisions_columns].values[0]
          , title2: df2[collisions_columns].values[0]
          , 'Results_Lower_is_Better': collisions_columns}) 
    
    final_df_r = pd.DataFrame({title1: df1[road_columns].values[0]
              , title2: df2[road_columns].values[0]
              , 'Results_Lower_is_Better': road_columns}) 
    
    final_df_o = pd.DataFrame({title1: df1[other_columns].values[0]
          , title2: df2[other_columns].values[0]
          , 'Results_Lower_is_Better': other_columns}) 
        
        
    final_df_s.index = final_df_s.Results_Higher_is_Better
    final_df_c.index = final_df_c.Results_Lower_is_Better
    final_df_o.index = final_df_o.Results_Lower_is_Better
    final_df_r.index = final_df_r.Results_Lower_is_Better
    return final_df_s, final_df_c, final_df_o, final_df_r

def show_values(g, pr="%", is_vertical = True):
    for p in g.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy() 
        if is_vertical:
            g.annotate(f'{round(height,1)}{pr}' if pr == "%" else f'{height}{pr}', (x + width/2, y + height*1.02), ha='center')
        else:
            g.annotate(f'{width}{pr}', (x + width + 0.09, y + height/2), ha='center')

```

```python
import matplotlib.pyplot as plt
font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 17}

plt.rc('font', **font)


final_df_s, final_df_c, final_df_o, final_df_r = mergeTwoLeaderbordMetricsResults('MLMAS1','MLMAS2', mlmas1_df ,mlmas2_df)
graph = final_df_s.plot.bar(figsize=(13,8))

graph2 = final_df_c.plot.bar(figsize=(13,8))
graph3 = final_df_r.plot.bar(figsize=(13,8))
graph4 = final_df_o.plot.bar(figsize=(11,9))

plt.tight_layout()
show_values(graph)
show_values(graph2, pr="")
show_values(graph3, pr="")
show_values(graph4, pr="")
```

```python
def get_driver_percentage():
    # remove all control with -1
    df_driver_new = df_driver[df_driver['control'] != "-1"]
    # find the overall percentage of each driver
    return df_driver_new['driver'].value_counts(normalize=True) * 100
```

```python
def get_jason_conditions_percentage():
    cond_p = {}
    df_driver_new = df_driver[df_driver['driver'] == "Jason"]
    df_driver_new = df_driver_new[df_driver_new['control'] != "-1"]
    # get all the unique values of column 'conditions'
    conditions = df_driver_new['condition'].unique()
    # remove nan from the list
    conditions = [x for x in conditions if str(x) != 'nan']
    # get the percentage of each condition
    for condition in conditions:
        cond_p[condition] = len(df_driver_new[df_driver_new['condition'] == condition]) / len(df_driver_new) * 100

    return cond_p
```

```python
def calculate_infraction_count(df):
    list_of_infractions = {}
    meta_data = {}
    result = []
    collisions_layout, collisions_pedestrian, collisions_vehicle, outside_route_lanes, red_light, route_dev, route_timeout, stop_infraction, vehicle_blocked = 0,0,0,0,0,0,0,0,0

    # read the json file
    with open(df) as f:
        data = json.load(f)
        main_list = data['_checkpoint']['records']

        for i in main_list:
            list_of_infractions[i['index']] = i['infractions']
            meta_data[i['index']] = i['meta']

    # 'collisions_vehicle'
    for routes in list_of_infractions.items():
        # print(routes)
        for inf in routes[1].items():
            if len(inf[1]) > 0:
                if inf[0] == 'collisions_vehicle':
                    collisions_vehicle = collisions_vehicle+len(inf[1])
                elif inf[0] == 'collisions_pedestrian':
                    collisions_pedestrian = collisions_pedestrian+len(inf[1])
                elif inf[0] == 'collisions_layout':
                    collisions_layout = collisions_layout+len(inf[1])
                elif inf[0] == 'outside_route_lanes':
                    outside_route_lanes = outside_route_lanes+len(inf[1])
                elif inf[0] == 'red_light':
                    # print(inf[1])
                    red_light = red_light+len(inf[1])
                    # print(red_light)
                elif inf[0] == 'route_timeout':
                    route_timeout = route_timeout+len(inf[1])
                elif inf[0] == 'route_dev':
                    route_dev = route_dev+len(inf[1])
                elif inf[0] == 'stop_infraction':
                    stop_infraction = stop_infraction+len(inf[1])
                elif inf[0] == 'vehicle_blocked':
                    vehicle_blocked = vehicle_blocked+len(inf[1])

    return [collisions_layout, collisions_pedestrian, collisions_vehicle, outside_route_lanes, red_light, route_dev, route_timeout, stop_infraction, vehicle_blocked]
```

```python
import matplotlib.pyplot as plt

# Create a list of the incidents
incidents = ["Collisions layout", "Collisions pedestrian", "Collisions vehicle", "Outside route lanes", "Red light", "Route dev", "Route timeout", "Stop infraction", "Vehicle blocked"]

# Create a list of the number of incidents for each type
incident_counts = calculate_infraction_count(hilal_mlmas)

# Find the highest and lowest incidents
highest_incident = incidents[incident_counts.index(max(incident_counts))]
lowest_incident = incidents[incident_counts.index(min(incident_counts))]

# Set the colors for the bars
highest_color = "red"
lowest_color = "blue"

# Create the pie chart
plt.pie(incident_counts, labels=incidents, colors=[highest_color if incident == highest_incident else lowest_color for incident in incidents], autopct='%1.1f%%')

# Add a title to the chart
plt.title("Hilal - Infraction Distribution")

# Show the chart
plt.show()
```
